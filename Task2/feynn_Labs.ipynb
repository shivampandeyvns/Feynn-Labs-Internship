# K-Means on Tamilnadu state diabetes

```python
df=pd.read_csv("hypertension_diabetes_patients_Treated_from_01.04.2018_to_31.03.2019.csv")
#################################################################################################################
df.rename(columns={'Name of the District':'District', 
                   'No. of Male Hypertension Patients':'Male',
                   'No. of Female Hypertension Patients':'Female'  , 
                   'Total No. of Hypertension Patients Treated':'Total_No_of_Patients_Treated_ht' , 
                   'No. of Male Diabetes Patients Treated':'Male_Treated_db' , 
                   'No. of Female Diabetes Patients Treated':'Female_Treated_db' ,
                   'Total No. of Diabetes Patients Treated': 'Total_No_of_Patients_Treated_db'}, inplace=True)
#################################################################################################################
df = df.drop(36)          
```

```python
df.tail(5)
```

```python
len(set(df["District"]))
```

```python
district=df["District"]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df["District"])
label=list(le.transform(df["District"]))
df["District"]=label
```

```python
df.tail(5)
```

```python
from sklearn.cluster import KMeans
```

```python
X = np.array(df)
kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
label=kmeans.labels_
```

```python
classes_state= pd.DataFrame({"district":district ,"label":label })
```

```python
classes_state
```

```python
sns.scatterplot(classes_state["label"] ,classes_state["district"] )
```

# K-Means on states -  diabetes

```python
df1=pd.read_csv("diabetes_India.csv")
##################################################################################################################################################
df1.rename(columns={'NFHS III(2005-06)-Number of women in age group 15-49 per 100-000 reported Diabetes-Total':'women_Diabetes_Total', 
                   'NFHS III(2005-06)-Number of women in age group 15-49 per 100-000 reported Diabetes-Rural':'women_Diabetes_Rural' ,
                   'NFHS III(2005-06)-Number of women in age group 15-49 per 100-000 reported Diabetes-Urban':'women_Diabetes_Urban' , 
                   'NFHS III(2005-06)-Number of men in age group 15-49 per 100-000 reported Diabetes-Total':'men_Diabetes_Total' , 
                   'NFHS III(2005-06)-Number of men in age group 15-49 per 100-000 reported Diabetes-Rural':'men_Diabetes_Rural' , 
                   'NFHS III(2005-06)-Number of men in age group 15-49 per 100-000 reported Diabetes-Urban':'men_Diabetes_Urban' }, inplace=True)
##################################################################################################################################################
df1 = df1.drop(0)
##################################################################################################################################################
states = df1["India/States/UTs"] 
##################################################################################################################################################
df1.head(5)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df1["Categories"])
label=list(le.transform(df1["Categories"]))
df1["Categories"]=label
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df1["India/States/UTs"])
label=list(le.transform(df1["India/States/UTs"]))
df1["India/States/UTs"]=label
```

```python
X = np.array(df1)
kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
kmeans.labels_
```

```python
interia_list =[]
for i in range(1,11):
    kmns= KMeans(n_clusters = i, init = 'k-means++')
    kmns.fit(X)
    interia_list.append(kmns.inertia_)   
```

```python
sns.set(rc={'figure.figsize':(10 , 7)})
plt.plot(range(1,11), interia_list)
plt.xlabel('Number of Clusters')
plt.ylabel('interia')
plt.title('Elbow Method')
plt.grid(True)
plt.show()
```

```python
kmeans= KMeans(n_clusters =4, init = 'k-means++')
kmeans.fit(X)
label=kmeans.labels_
```

```python
india_state= pd.DataFrame({"states":states ,"label":label })
```
